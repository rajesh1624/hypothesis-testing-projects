### Pista House selling Hyderabadi biryani claims that each parcel packet has 500 grams of biryani (also mentioned on the packet label).You are sceptical of their claim and believe that, on average, each packet does not contain 500 grams of biryani. How do you prove your claim? (Given that the population standard deviation is 50.)

```python
# Create a function to calculate the z_score
```

```python
def z_score(sample_mean, pop_mean, pop_std, sample_size):
    a = sample_mean - pop_mean
    b =  pop_std/sample_size ** 0.5
    return a/b
```

```python
# Each biryani packet weighs 500g on average.So we suspect itâ€™s not exactly 500g.So this is a two-tailed Z-test.
```

```python
# Sample data
l = [490, 220, 470, 500, 495, 496, 496, 498, 508, 480]
```

```python
sum(l)
```

```python
len(l)
```

```python
sample_mean =  sum(l)/ len(l)
sample_mean
```

```python
# Sample size and sample parameter

sample_mean = 465.3
sample_size = 10
```

```python
# Poputaion parameters

pop_mean = 500
pop_std = 50
```

```python
# We will first calculate the z_value

z_value =  z_score(sample_mean, pop_mean, pop_std, sample_size)
z_value
```

```python
# We will first calculate the left tail test using norm function.
```

```python
# First way 

p_value = norm.cdf(z_value, 0, 1)
p_value 
```

```python
# Second Way
```

```python
sample_std_dev = pop_std/sample_size ** 0.5
sample_std_dev
```

```python
p_value = norm.cdf(sample_mean,pop_mean,sample_std_dev)
p_value
```

```python
p_value = norm.cdf(sample_mean,pop_mean, pop_std/(10**0.5))
p_value             
```

```python
# Conslusion using p-value
```

```python
alpha = 0.05
if p_value < alpha:
    print('Reject Null Hypothesis')
else:
    print('Fail to Reject Nulll Hypothesis')
```

# Question 2: Hypothesis Testing

### You have developed a new Natural Language Processing (NLP) Algorithm and conducted a user study. You claim that the average rating given by users is greater than 4 on a scale of 1 to 5.How do you prove this claim to your client using hypothesis testing?(Assume that the population standard deviation is 0.5.)

```python
l = [5,2,3,4,5,4,3,5,4,3,4,4.5,2,4,3,3,4,5,4,5]

sum(l)
print(len(l))

sample_mean = sum(l)/len(l)
sample_mean
```

```python
# Framing Null and Alternate hypothesis
```

## H0: The average rating is 4.
## H1: The average rating is less than 4

```python
# sample parameters

sample_size = len(l)
sample_mean = sample_mean
```

```python
# population parameter

pop_mean = 4
pop_std = 0.5
```

```python
# Calculate z_score
```

```python
z_value = z_score(sample_mean, pop_mean, pop_std, sample_size)
z_value
```

```python
# Calculate p-value
```

```python
# First way
p_value =  norm.cdf(z_value)
p_value
```

```python
# Second Way
```

```python
p_value = norm.cdf(sample_mean, pop_mean,pop_std/(20**0.5))
p_value
```

```python
alpha = 0.05
if p_value < alpha:
    print('Reject Null Hypothesis')
else:
    print('Fail to Reject Nulll Hypothesis')
```
